### Explanation of the export code

`summons_sample` is the dataframe we want to export

`.to_csv()` saves our dataframe to a .csv file. You are baseically exporting your python data out into a format that Excel, Google Sheets can read.

`index=False`
By default, on export pandas includes the DataFrame's index – the row numbers (0, 1, 2, 3...). We NEVER
need it 99.99 percent of the time. `index=False` simply tells pandas to leave that column out on export.

`encoding="utf-8"`
Tells Python which character encoding to use when writing the file. `UTF-8` is the most universal standard — it handles English plus accented characters, symbols, non-Latin scripts, etc. without garbling them. If your data has names like "García" or "Müller" and you skip this, you might get weird characters or errors depending on your system.

### Prompt: 

Provide the pandas code to calculate which boroughs had the most summons (in descending order).


````python
# 1. Which borough had the most summons? 

# Replace string '(null)' placeholders with real NaN so they're excluded cleanly
nypd_summons['BORO'] = nypd_summons['BORO'].replace('(null)', pd.NA)

# Count summons per borough, sort descending, and drop any missing borough values
borough_counts = (
    nypd_summons['BORO']
    .value_counts()              # Count occurrences of each borough; sorts descending by default
    .reset_index()               # Convert the Series into a two-column DataFrame
    .rename(columns={            # Rename columns from pandas defaults to readable names
        'BORO': 'borough',
        'count': 'summons_count'
    })
)

borough_counts
````

````python
## print the columns
print(nypd_summons.columns.tolist())
````

````python
## define offen

offen = 'OFFENSE_DESCRIPTION'
````

````python
# 2. What were the top 10 most common offenses?
nypd_summons[offen] = nypd_summons[offen].replace('(null)', pd.NA)

offense_descriptions = (
    nypd_summons[offen]
    .value_counts()             
    .reset_index()
    .head(10)  
    .rename(columns={            
        offen: 'OFFENSE_DESCRIPTION',
        'count': 'summons_count'
     })  
) 
print(offense_descriptions) 
````

````python
# 3. What are the total number of summons broken down by year?
nypd_summons ['YEAR'] =  pd.to_datetime(nypd_summons['YEAR'])

summons_by_year = (
nypd_summons.groupby('YEAR')
.size()
.reset_index(name='summons_count')
)
print(summons_by_year)
````

````python
##define unusual summons/rare offenses
rare_offenses = 'OFFENSE_DESCRIPTION'
````

````python
# 4. What were some of the most unusual summons?
rare_offenses = (
    nypd_summons[offen]
    .value_counts()
    .reset_index()
    .tail(10)
    .rename(columns={offen: 'OFFENSE_DESCRIPTION', 'count': 'summons_count'})
)

print(rare_offenses)
````

````python
# 5. What is the count of summons by gender?

##define gender summons

gender = 'SEX'

gender = (
    nypd_summons[offen]
    .value_counts()
    .reset_index()
    .rename(columns={offen: 'SEX', 'count': 'summons_count'})
)

print(gender)
````

````python
# 6. How many summons were issued by race?

##define race summons

race_summons = 'RACE'

race = (
    nypd_summons[offen]
    .value_counts()
    .reset_index()
    .rename(columns={offen: 'RACE', 'count': 'summons_count'})
)

print(race)
````

````python
#7. Which borough had the fewest offenses per 100K people?


# 2020 Census population
populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}
````

````python
## answer to above 

##define summons by borough
summons_by_borough = (
    nypd_summons.groupby('BORO')
    .size()
    .reset_index(name='summons_count')
        )

#add population and calculate per 100K
summons_by_borough['population'] = summons_by_borough['BORO'].map(populations)
summons_by_borough['per_100k'] = (
    summons_by_borough['summons_count'] / summons_by_borough['population'] * 100000
).round(2)

## sort to find fewest
summons_by_borough = summons_by_borough.sort_values('per_100k')

print(summons_by_borough)

summons_by_borough = summons_by_borough.sort_values('per_100k')

display(summons_by_borough)

fewest = summons_by_borough.iloc[0]
print(f"{fewest['BORO']} had the fewest offenses per 100K people ({fewest['per_100k']})")
````
