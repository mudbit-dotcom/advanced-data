### How many values in a column?



## Counting

-  ```sum``` adds up all the values in a dataset, to give you the total amount.

-  ```count``` tells you how many items of particular values exist in a dataset. It doesn't care about the values themselves, just the number of entries.

So if you have the following of donations by individuals:

| Donor | Amt   |
|-------|-------|
| A     | 200   |
| A     | 200   |
| B     | 300   |
| B     | 100   |
| A     | 200   |
| A     | 100   |


- The ```sum``` of donations is: 1,100
- The ```count``` of donations is: 6

The ```count``` of donors is:
- ```A appears 4```
- ```B appears 2```

The count of amounts:
- ```contribs of 200 were given 3 times```,
- ```contribs of 100 were given 2 times```,
- ```contribs of 300 were given 1 times```



We use the ```.value_counts()``` method:

```df_name["column_name"].value_counts()```



````python
## bring in our diamonds dataset
diamonds_df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/diamonds.csv")
diamonds_df
````

````python
## number of unique cuts
diamonds_df["diamond cut"].nunique()
````

````python
diamonds_df["diamond cut"].value_counts(ascending = True)
````

````python
diamonds_df["diamond cut"].value_counts(ascending = False).to_frame("Quantity")
````

````python
## what is the break down diamond_cut
diamonds_df["diamond cut"].value_counts().to_frame("Quantity").sort_values(by="Quantity", ascending = True)
````

### ```value_counts()``` parameter

By default, it is ```.value_counts(normalize = False)```

This gives you want we calculated above

But ```.value_counts(normalize = True)``` returns  proportions (fractions of the total).

````python
## get as proportion
diamonds_df["diamond cut"].value_counts(normalize = True)
````

## Round it

```round(item_to_round, number_of_decimal_places)```

````python
## return as percentage, rounded 2 decimal place
round(diamonds_df["diamond cut"].value_counts(normalize = True) * 100, 2)
````
